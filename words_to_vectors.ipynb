```python
import pandas as pd
df = pd.read_csv('spam.csv',encoding='ISO-8859-1')
df.head()
```

```python
# Data Cleaning and Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

porter_stemmer = PorterStemmer()
```

```python
df = df[['Label','Message']]
df.head()
```

```python
corpus = []
for i in range(0,len(df)):
    review = re.sub('[^a-zA-z]',' ',df['Message'][i])
    review = review.lower()
    review = review.split()
    review = [porter_stemmer.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Creating bag of words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)  # for binary bow, enable 'binary=True'
```

```python
x = cv.fit_transform(corpus).toarray()
x
```

```python
# For creating N Gram, we will just use an additional parameter in the CountVectorizer function
from sklearn.feature_extraction.text import CountVectorizer
cv2 = CountVectorizer(max_features=1000,ngram_range=(2,3))  # ngram_range=(1,1) will be same as bow
```

```python
x2 = cv2.fit_transform(corpus).toarray()
```

```python
cv2.vocabulary_ # We passed (2,3) in the function parameter therefore we got combination of 2 and 3 words.
# To imoprove the efficiency, we can play around these hyperparameters.
```

#### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=250)
x3 = tfidf.fit_transform(corpus).toarray()
```
